```python
with open("/plant/predict2/labels/plant.txt", "r") as f:
   words = [line.split()[0] for line in f.readlines()]
```

```python

mid_d =list(set(words))
mid_d = list(map(int, mid_d))
get_d =  {k: all_d[k] for k in mid_d if k in all_d}
kinds=[0,7,15,22,29,36]
stb={1:'딸기잿빛곰팡이병',2:'딸기흰가루병',3:'냉해피해',4:'다량원소결핍(N)',
5:'다량원소결핍(P)',6:'다량원소결핍(K)'}
tmt={8:'토마토흰가루병',9:'토마토잿빛곰팡이병',
10:'열과',11:'칼슘결핍',12:'다량원소결핍(N)',13:'다량원소결핍(P)',14:'다량원소결핍(K)'}
pap={15:'파프리카',16:'파프리카흰가루병',17:'파프리카잘록병',18:'칼슘결핍',19:'다량원소결핍(N)',20:'다량원소결핍(P)',
21:'다량원소결핍(K)'}
cuc={23:'오이노균병',24:'오이흰가루병',25:'냉해피해',26:'다량원소결핍(N)',
27:'다량원소결핍(P)',28:'다량원소결핍(K)'}
pep={30:'고추탄저병',31:'고추흰가루병',32:'칼슘결핍',
33:'다량원소결핍(N)',34:'다량원소결핍(P)',35:'다량원소결핍(K)'}
grp={37:'시설포도탄저병',38:'시설포도노균병',
39:'일소피해',40:'축과병'}

for key, value in get_d.items():
    if key in kinds:
      if key == 0:
          kind = stb
      elif key == 7:
          kind = tmt
      elif key == 15:
          kind = pap
      elif key == 22:
          kind = cuc
      elif key == 29:
          kind = pep
      elif key == 36:
          kind = grp
      else:
        print('이미지오류')

      for key1, value1 in kind.items():
          if key1 in list(get_d.keys()):
            print('품종='+' '+value+'\n'+'병='+' '+value1 )

          else:
            print('품종='+' '+value)
            break
```
